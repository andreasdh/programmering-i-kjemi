# Differensiallikninger

```python
import matplotlib.pyplot as plt
import numpy as np

# Initialbetingelser
x0 = 0
x_slutt = 10
dx = 1E-3
N = int((x_slutt - x0)/dx) + 1
y0 = [0,1,2,3,4]

# Differensiallikningen
def fder(y, x):
    return y + x/5

# Arrayer
x = np.zeros(N)
y = np.zeros(N)

x[0] = x0

# Eulers metode
for startverdi in y0:
    y[0] = startverdi
    for i in range(N-1):
        y[i+1] = y[i] + fder(y[i],x[i])*dx
        x[i+1] = x[i] + dx
    plt.plot(x,y,label=f"y0 = {startverdi}")

plt.legend()
plt.grid()
plt.show()
```

## Newtons 2. lov

```python
# Initialbetingelser
m = 1  # kg
k = 0.1 # luftmotstandskoeffisienten
v0 = 0  # startfart i m/s
s0 = 10 # startposisjon i m
g = 9.8 # tyngdeakselerasjon i m/s^2

# Tidssteg
dt = 1E-3
t0 = 0
t_slutt = 2 # tid i sekunder
N = int((t_slutt-t0)/dt) + 1

# Arrayer
t = np.zeros(N)
s = np.zeros(N)
v = np.zeros(N)
a = np.zeros(N)

s[0] = s0
v[0] = v0

for i in range(N-1):
    a[i] = - g - k*v[i]/m
    # Eulers metode
    v[i+1] = v[i] + a[i]*dt
    s[i+1] = s[i] + v[i]*dt
    t[i+1] = t[i] + dt
```

```python
plt.plot(t, s)
plt.show()
```
