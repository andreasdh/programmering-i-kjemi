# Programmering i kjemi

Det finnes mange biblioteker i Python som kan brukes til å utforske sammenhenger i kjemi. Tre nyttige biblioteker er _mendeleev_, _chemlib_ og _pHcalc_. De kan installeres enkelt hvis du har installert Anaconda-pakka. Gjør følgende:

1. Åpne "Anaconda Prompt" på maskinen din. Dette gir et svart terminalvindu som du kan skrive i.
2. Skriv "pip install mendeleev" og trykk enter. Det er mulig du må bekrefte dette ved å trykke "y" og enter dersom dette spørres om.
3. Da skal _mendeleev_ være installert. Gjør det samme med "pip install pHcalc" og "pip install chemlib".

Her er tre eksempler på hvordan bibliotekene kan brukes:

```python
from pylab import *
from mendeleev import element

atomnummer = []
elektronegativitet = []
gruppe1 = [1, 3, 11, 19, 37, 55, 87]

for i in gruppe1:
    grunnstoff = element(i)
    atomnummer.append(grunnstoff.atomic_number)
    elektronegativitet.append(grunnstoff.electronegativity())
    
plot(atomnummer, elektronegativitet, linestyle=" ", marker="o")
xlabel("Atomnummer")
ylabel("Elektronegativitet")
show()
```

```python
from chemlib import Reaction, Compound
# Definerer forbindelser
H2 = Compound("H2")
I2 = Compound("I2")
HI = Compound("HI")

# Definerer reaksjon som en liste med reaktanter og en med produkter
reaksjon = Reaction([H2, I2], [HI])
print(reaksjon.formula)     # Printer reaksjonslikningen
print(reaksjon.is_balanced) # Sjekker om reaksjonen er balansert

reaksjon.balance()          # Balanserer reaksjonen
print(reaksjon.formula)
print(reaksjon.is_balanced)
```

```python
from pylab import *
from pHcalc.pHcalc import Acid, Neutral, System

# pH i  0,01 M KF
hf = Acid(Ka=6.76E-4, charge=0, conc=0.01)
k = Neutral(charge=1, conc=0.01)
system1 = System(hf, k)
system1.pHsolve()
print(system1.pH)
```

## Forslag til oppgaver
Ikke vær redd for å gi elevene store deler av, eller hele, programmer. Poenget er å utforske kjemien, ikke bli gode til å programmere. Et eksempel på en oppgave kan være:

1. Beskriv hva programmet nedenfor gjør. 
2. Kjør deretter programmet og beskriv plottet.
3. Lag samme type plott som lages i oppgave 1, men med karbonsyre istedenfor fosforsyre. Diskuter likheter og forskjeller i plottene.

```python
from pylab import *
from pHcalc.pHcalc import Acid, Neutral, System

H3PO4 = Acid(pKa=[2.15, 7.20, 12.32], charge=0, conc = 0.01)
ph = linspace(0, 14, 10000)
andel = H3PO4.alpha(ph)

plot(ph, andel)
legend(["H$_3$PO$_4$", "H$_2$PO$_4^-$", "HPO$_4^{2-}$", "PO$_4^{3-}$"])
xlabel("pH")
ylabel("Prosentvis fordeling")
show()
```

## Datahåndtering
Vi kan håndtere data på en enkel måte ved hjelp av enten pylab-biblioteket eller Pandas. Her er en måte å gjøre det på med Pandas:

```python
from pylab import *
import pandas as pd

data = pd.read_csv("titreringsdata.txt", delimiter = ",")
scatter(data["volum"], data["pH"], color='seagreen')
xlabel("Volum tilsatt NaOH")
```

Fila "titreringsdata.txt" må ligge i samme mappe som programmet ditt for at dette skal fungere.
