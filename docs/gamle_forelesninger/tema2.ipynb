```python
import numpy as np
import matplotlib.pyplot as plt
```

## Titrering

```python
data = np.loadtxt('Datafiler-innhold/titrering.txt', skiprows = 2, delimiter = ',')

volum = data[:,0]
pH = data[:,1]

plt.plot(volum,pH,marker='.')
plt.title('Titrering av eddiksyre')
plt.xlabel('Volum tilsatt NaOH (mL)')
plt.ylabel('pH')
plt.axvline(x=31.5)
plt.axhline(y=9)
plt.grid()
plt.show()
```

```python
# Interpolasjon
from scipy.interpolate import interp1d

T = [0, 20, 40, 60, 80, 100]
sol = [88.5, 56, 34, 20, 11, 7]

f = interp1d(T, sol, kind = 'linear')
x_new = np.linspace(T[0], T[-1], 10000)

y = f(x_new)

plt.plot(T,sol,linestyle='',marker='*',color='gold',label='Datapunkter')
plt.plot(x_new,y,label='Interpolerte data')
plt.legend()
plt.show()
```

```python
f(50)
```

```python
jern = np.array([0, 0.010, 0.030, 0.060, 0.080, 0.10]) # Konsentrasjon i mol/L
absorbans = np.array([0, 0.021, 0.120, 0.139, 0.186, 0.208]) # Absorbans

a, b = np.polyfit(jern,absorbans,1)

x = np.linspace(0, 0.15, 10000)
y = a*x + b

plt.plot(jern,absorbans,linestyle='',marker='*',color='firebrick',label='Datapunkter')
plt.plot(x,y,label='Tilpasset funksjon')
plt.legend()
plt.axhline(y=0.25,color='red')
plt.axvline(x=0.11335640138408302,color='red')
plt.show()
```
